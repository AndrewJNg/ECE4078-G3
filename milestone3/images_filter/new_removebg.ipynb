```python
# Import modules
from PIL import Image
from rembg import remove
import io
import glob
import os
```

```python
# Main function
def remove_bg_batch(processApple, processCapsicium, processGreenApple, processMango, processOrange, rename, deleteExistingFiles):
    fruitName = ""
    for iter in range(1,6):
        if iter == 1 and processApple: # apple
            fruitName = "apple"
            inputDirectory = "1_taken_pictures/apple/"
            outputDirectory = "2_removedbg/apple/"
        elif iter == 2 and processGreenApple: # greenapple
            fruitName = "greenapple"
            inputDirectory = "1_taken_pictures/greenapple/"
            outputDirectory = "2_removedbg/greenapple/"
        elif iter == 3 and processOrange: # orange
            fruitName = "orange"
            inputDirectory = "1_taken_pictures/orange/"
            outputDirectory = "2_removedbg/orange/"
        elif iter == 4 and processMango: # mango
            fruitName = "mango"
            inputDirectory = "1_taken_pictures/mango/"
            outputDirectory = "2_removedbg/mango/"
        elif iter == 5 and processCapsicium: # capsicum
            fruitName = "capsicium"
            inputDirectory = "1_taken_pictures/capsicium/"
            outputDirectory = "2_removedbg/capsicium/"
        else:
            continue

        if not os.path.exists(outputDirectory):
            os.makedirs(outputDirectory)
        else:
            if(deleteExistingFiles == 1):
                remove_files_in_folder(outputDirectory)

        fileCount = 1
        for fileName in os.listdir(inputDirectory):
            if fileName.endswith(".png") or fileName.endswith(".jpg"):
                imgPath = os.path.join(inputDirectory, fileName)
                
                baseFileName, extension = os.path.splitext(fileName)
                extension = '.png'
                if rename:
                    saveFileName = os.path.join(outputDirectory, f"{fruitName}_nobg_{fileCount}{extension}")
                else:
                    saveFileName = os.path.join(outputDirectory, f"{baseFileName}_nobg{extension}")
                
                input = Image.open(imgPath)
                output = remove(input)
                output.save(saveFileName)
                
                print("Saved", saveFileName)
                fileCount += 1
        print("Removed background for pics in ", inputDirectory, "\n")

def remove_files_in_folder(folderPath):
    try:
        # List all files in the folder
        files = os.listdir(folderPath)

        # Loop through the files and remove them
        for file in files:
            filePath = os.path.join(folderPath, file)
            if os.path.isfile(filePath):
                os.remove(filePath)
                #print(f"Removed: {filePath}")

        print("Removed files in ", folderPath)

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
# Run function
# IMPORTANT
# process___ : removes background for selected fruits folder
# rename: takes randomPicName.png and creates a no_bg image with name fruit_nobg_1.png
# deleteExistingFiles: deletes the output directory before creating images into them

remove_bg_batch(processApple = 1,
            processCapsicium = 1,
            processGreenApple = 1,
            processMango = 1,
            processOrange = 1,
            rename = 0,
            deleteExistingFiles = 1) # <- IMPORTANT
```
